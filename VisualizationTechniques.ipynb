````r
# Exponentially xponential Distribution Curve with parameter 1/theta
theta <- as.numeric(readline(prompt = "Enter the value of theta: "))

if (theta <= 0) {
  stop("Theta must be a positive number!")
  }

  # Create a sequence of x values
  x <- seq(0, 10, by = 0.01)

  # Exponential PDF formula: f(x) = (1/theta) * exp(-x/theta)
  pdf_values <- (1 / theta) * exp(-x / theta)

  # Plotting the exponential distribution
  plot(
    x, pdf_values, type = "l", col = "blue", lwd = 2,
      xlab = "x", ylab = "Density",
        main = paste("Exponential Distribution (theta =", theta, ")")
        )

        grid(col = "gray")
````

# Advanced Visualization `Contour Plot`
```
using contour(x, y, z, nlevels= , levels= ,labels= , xlim= ,ylim= , zlim= , col= )
* x, y are locations of grid lines at which the value measured in z are measured.
* z is amatrix containing the valuesto be plotted.
* nlevels has the number of contour levels desired if levels is not supplied.
* levels is numeric vector of levels at which to draw contour lines.
* labels is a vector giving the labels for the contour lines. if NULL then the levels are used as labels.
* xlim , ylim, zlim have the limits of x, y , z-axes.
* col is for the color of the lines

```


````r
matrixData <- matrix(
    c(12,32,3,23,32,45,34,56,43), nrow=3, ncol=3
)
contour(
    x=seq(0,1,length.out = nrow(matrixData)),
    y=seq(0,1,length.out = ncol(matrixData)),
    matrixData,
    xlab="x axis",
    ylab="y axis",
    col="blue"
)
````
