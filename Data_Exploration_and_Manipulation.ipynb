# Using Basic Statistical/Mathematical Functions to Explore Missing Values using  **`na.rm = TRUE or FALSE`**

````r
mean(
    airquality[,1], na.rm = TRUE
)
mean(
    airquality$Solar.R
)
cor(airquality$Solar.R, airquality$Month)
# they contain some mising values so the RESULT is NA. the missing can be removed by function "na.rm = TRUE or FALSE"
````

# Using Basic Statistical/Mathematical Functions to Explore Missing Values using **`na.omit()`**

````r
newAirquality <- na.omit(airquality)
print(newAirquality)
par(mfrow = c(1,3))
aggr(
    newAirquality, numbers = FALSE, prop = FALSE
)
matrixplot(newAirquality)
marginplot(
    newAirquality[c("Ozone", "Solar.R")], pch = c(18), col = c("green", "red", "blue")
)
cor(newAirquality$Ozone, newAirquality$Solar.R)
#This function returns the object with list wise deletion of missing values it deletes all those records where the value is NA.
````

2.1 Data Melting
```
we can melt the data to organize it converting all columns ther than species into multiple rows.
```

````r
iris          # a built in dataset
install.packages("reshape2")
library(reshape2)
molten.iris <- melt(iris, id = c("Species"))
print(molten.iris)
# Species is the only column ehci has a categorical value. there are only three categorical values for species namely setosa, versicolor and virginica. the command "molten.iris <- melt(iris, id = c("Species"))" melts the data on the basis of species categorical variable.
````

2.2 Casting the Molten Data

````r
cast.iris <- dcast(
    molten.iris,
    Species~variable,
    sum
)
print(cast.iris)
plot(cast.iris)   #additionally
# the "dcast()" command specifies that "Species" is the variable (Species~Variable) and the sum is calculated for all data columns. the result displays three categories of species in the iris dataset and sum of the rest of the variables. thus sum of sepal.length, sepal.width, petal.length and petal.width is calculated for each of the three species - setosa, versicolor and virginica.
````

3.1 The **`apply`**
```
the apply() function performs an operations on each part of data management. it returns vector, array , list , matrix obtained by applying
a function to margins of an array or matrix. it is used to evaluate a function over margins of an array. it is most often used to apply a function to all the rows or columns of a matrix or array.
syntax : apply(x, v, func....)
         * x is an array including a matrix
         * v is a vector generally specifying the subcripts on which the
          function will be applied over. In a matrix "1" indicates rows and "2" indicates columns. the vector c(1,2) indicates both rows and columns.
         *func is the function to be applied including sum, mean and median.

```

````r
# @title apply() function
state <- as.data.frame(state.x77, stringsAsFactors = FALSE)
#print(state)
x1 <- state$Population
x2 <- state$Income
x3 <- state$Illiteracy
x<- cbind(x1,x2,x3)
#print(x)
apply(x,2, sum)
apply(x,1,sum)
````

the **`sapply()`** function


```
the sapply() function is similar to lapply() functions, where the function is applicable to every element in the a list.
the difference is in the return value. unlike lapply() where the result is stored in list form, sapply() stores the result in the form of a vector or matrix. besides sapply() will try to simplify the result of lapply() if possible. essentially sapply() calls lapply() on its input and the result is generated according to the input. if
the list result is a list where every element is of length 1, then a vector is returned, if the result is a list where every element is
a vector of the same length (>1), a matrix is returned and if it can`t figure things out a list is returned.
```



````r
sapply(list1, summary)
typeof(sapply(list1, summary))
````

```
Another widely used variant of the apply() function is
the tapply() function. The tapply() function is used to apply a function on each
set broken up by a given set of factors. This function will take a list of data
(usually a vector), a list of factors of the same list, and a function. It will
then apply the function
 to each subset of the data that matches each of the factors. It can be thought
 of as a combination of split() and sapply() functions for vectors only.
```



````r
result <- data.frame(
    marks = c(11,22,10,45,34,33),
    city = as.vector(c("Mumbai", "Pune", "Mumbai", "Mumbai","Mumbai","Gudgoan"))
)
print(result)
result2 <- tapply(result$marks, result$city, mean)
print(result2)
result3 <- tapply(result$marks, result$city, summary)
print(result3)
plot(result2, type = "l")
result3$Mumbai   #FOR INDIVIDUALLY
````

Data frame is a key data structure in statistics and in R. The basic structure
of a data frame is that there is one observation per row and each column
represents a variable or characteristic of that observation. All the data

In the Chapter 2, we have already discussed some tools like the subset() function, [ and $ operators to extract subsets of data frames. Though R has an internal implementation of data frames that is used often and "dplyr" package does not provide any "new" functionality to R (that is, everything "dplyr" does can be done with base R), but it greatly simplifies and speeds up data frame management code and existing functionality in R.
